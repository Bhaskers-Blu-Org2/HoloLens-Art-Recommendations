# Find the K Nearest Neighbors for a particular image

here we find the k nearest neighbors based on image features. we load the pickles dictionary and find the nearest neighbors from this (which has the feature vectors for each image in the current dataset)

```python
import sys
sys.path.append("../main/")
```

```python
from image_utils import find_k_closest, get_k_nearest_neighbor_image
%load_ext autoreload
%autoreload 2

import random
import matplotlib.pyplot as plt
%matplotlib inline
```

- load the pickle file with feature vectors

```python
import pickle
with open('../data/features_dict.pickle', 'rb') as handle:
    features_dict = pickle.load(handle)
```

- here we show an example of finding the nearest neighbors for one particular image

```python
objectid = 676
this_vec = features_dict[objectid]
k_closest_list = find_k_closest(this_vec, features_dict)
image = get_k_nearest_neighbor_image(k_closest_list)
plt.imshow(image)
plt.show()
```

- here we show an example of finding the k nearest neighbors for a few randomly selected images

```python
features_dict_as_list = list(features_dict.items())

for _ in range(5):
    objectid, this_vec = random.choice(features_dict_as_list)
    k_closest_list = find_k_closest(this_vec, features_dict)
    image = get_k_nearest_neighbor_image(k_closest_list)
    plt.imshow(image)
    plt.show()
```
